```python
import pandas as pd


def convertervalores(df,colunas):
    for coluna in colunas:
        df[coluna] = df[coluna].str.replace(',','.').astype(float)
        df[coluna] =  df[coluna].round(2)
    df['Ano base'] = df['Ano base'].astype(int)
    df = df.drop_duplicates()
    return df

```

```python
Producao_Bruta = pd.read_csv('../Bronze/Producao_Bruta.csv',encoding='latin')
colunas = [
    "Quantidade Produção - Minério ROM (t)",
    "Quantidade Contido",
    "Quantidade Venda (t)",
    "Valor Venda (R$)",
    "Quantidade Transformação / Consumo / Utilização (t)",
    "Valor Transformação / Consumo / Utilização nesta mina (R$)",
    "Quantidade Transferência para Transformação / Utilização / Consumo (t)",
    "Valor Transferência para Transformação / Utilização / Consumo (R$)"
]
Producao_Bruta = convertervalores(Producao_Bruta,colunas)
Producao_Bruta.head()
```

```python
Producao_Beneficiada = pd.read_csv('../Bronze/Producao_Beneficiada.csv',encoding='latin')
colunas = [
    "Quantidade Produção",
    "Quantidade Contido",
    "Quantidade Venda",
    "Valor Venda (R$)",
    "Valor Consumo / Utilização na Usina (R$)",
    "Quantidade Transferência para Transformação / Utilização / Consumo",
    "Valor Transferência para Transformação / Utilização / Consumo (R$)"
]
Producao_Beneficiada = convertervalores(Producao_Beneficiada,colunas)
Producao_Beneficiada.head()
```

```python
Agua_Mineral_Producao = pd.read_csv('../Bronze/Agua_Mineral_Producao.csv',encoding='latin')
colunas = [
    "Quantidade (litros) Garrafão",
    "Quantidade (litros) Garrafa Plástica",
    "Quantidade (litros) Garrafa de Vidro",
    "Quantidade (litros) Copo",
    "Quantidade (litros) Outras Embalagens",
    "Valor (R$) Garrafão",
    "Valor (R$) Garrafa Plástica",
    "Valor (R$) Garrafa de Vidro",
    "Valor (R$) Copo",
    "Valor (R$) Outras Embalagens",
    "Quantidade (litros) Composição Produtos Industrializados",
    "Valor (R$) Composição Produtos Industrializados"
]
Agua_Mineral_Producao = convertervalores(Agua_Mineral_Producao,colunas)
Agua_Mineral_Producao.head()
```

```python
Agua_Mineral_Producao.to_csv('../Prata/Agua_Mineral_Producao.csv',index=False,encoding='latin')
Producao_Beneficiada.to_csv('../Prata/Producao_Beneficiada.csv',index=False,encoding='latin')
Producao_Bruta.to_csv('../Prata/Producao_Bruta.csv',index=False,encoding='latin')
```
